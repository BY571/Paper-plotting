```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import json
import glob
```

```python
# in each directory there have to be all the runs depending on the algorithm
data_1 = glob.glob("runs_JSON/SAC_ERE_PER/*.json")
data_2 = glob.glob("runs_JSON/SAC_PER/*.json")
data_3 = glob.glob("runs_JSON/SAC/*.json")
data_names = [data_1, data_2, data_3] 
```

```python
# !! All final must have the same length !!
prep_data1, prep_data2, prep_data3 = [],[],[]

for name in data_1:
    with open(name) as json_file:
        data_i = json.load(json_file)
    data_i = np.array(data_i)
    prep_data1.append(list(data_i[:,2]))
for name in data_2:
    with open(name) as json_file:
        data_i = json.load(json_file)
    data_i = np.array(data_i)
    prep_data2.append(list(data_i[:,2]))
for name in data_3:
    with open(name) as json_file:
        data_i = json.load(json_file)
    data_i = np.array(data_i)
    prep_data3.append(list(data_i[:,2]))
```

```python
fig = plt.figure(figsize=(10,7))
xdata = np.arange(len(prep_data1[0]))
data = (prep_data1, prep_data2, prep_data3)
linestyle = ['-.', '--', ':']
color = ['r', 'g', 'b']
label = ['SAC_ERE_PER', 'SAC_PER', 'SAC']
sns.set(style="darkgrid", font_scale=1.5)
for i in range(3):    
    sns.tsplot(time=xdata, data=data[i], color=color[i], linestyle=linestyle[i], condition=label[i])

plt.ylabel("Reward", fontsize=12)
plt.xlabel("Episodes", fontsize=12)
plt.title("Pendulum Env", fontsize=14)

plt.legend(loc='lower right')
#plt.savefig("GA_with_5_seeds_run.jpg", dpi=300)
plt.show()
```
